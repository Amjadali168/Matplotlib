# `Scatter Plot`

```python
import matplotlib.pyplot as plt
```

```python
day = [12,28,31,8,51]
no = [1,2,3,4,5]
plt.title('Scatter plot',fontsize=15)
plt.scatter(day,no,color="y")
plt.xlabel('Day',fontsize=12)
plt.ylabel('No.',fontsize=12)
plt.show()
```

```python
day = [12,28,31,8,51]
no = [1,2,3,4,5]
size = [40,300,200,100,150]
color = [10,20,30,40,50]
plt.title('Scatter plot',fontsize=15)
plt.scatter(day,no,c=color,s=size,cmap="viridis")
plt.colorbar()
plt.xlabel('Day',fontsize=12)
plt.ylabel('No.',fontsize=12)
plt.show()
```

```python
sepal_length = "Iris-setosa"
sepal_width = "Iris-virginica"
plt.scatter(sepal_length, sepal_width)
plt.show()
```
