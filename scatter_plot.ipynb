# `Scatter Plot`

```python
import matplotlib.pyplot as plt
```

```python
day = [12,28,31,8,51]
no = [1,2,3,4,5]
plt.title('Scatter plot',fontsize=15)
plt.scatter(day,no,color="y")
plt.xlabel('Day',fontsize=12)
plt.ylabel('No.',fontsize=12)
plt.show()
```

```python
day = [12,28,31,8,51]
no = [1,2,3,4,5]
size = [40,300,200,100,150]
plt.title('Scatter plot',fontsize=15)
plt.scatter(day,no,color="y",s=size)
plt.xlabel('Day',fontsize=12)
plt.ylabel('No.',fontsize=12)
plt.show()
```
