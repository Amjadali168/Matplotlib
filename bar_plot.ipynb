# `Bar plot`

```python
import matplotlib.pyplot as plt
```

```python
x = ['Python','C','Java','C++']
y = [85,70,80,60]
plt.title('Matplotlib')
plt.xlabel('Language',fontsize=10)
plt.ylabel('Popularity',fontsize=15)
c = ["r",'y','b','g']
plt.bar(x,y,color=c,width=0.4,edgecolor="black",linewidth=4)
plt.show()
```
