# `Bar plot`

```python
import matplotlib.pyplot as plt
```

```python
x = ['Python','C','Java','C++']
y = [85,70,80,60]
plt.title('Matplotlib')
plt.xlabel('Language',fontsize=10)
plt.ylabel('Popularity',fontsize=15)
c = ["r",'y','b','g']
plt.bar(x,y,color=c,width=0.4,edgecolor="black",linewidth=2,linestyle=":",alpha=0.4)   # (linestyle=":" is used to make the linewidth dotted)  (alpha is used to make the color blur)
plt.show()
```

```python
x = ['Python','C','Java','C++']
y = [85,70,80,60]
z = [20,40,50,55]
plt.title('Matplotlib')
plt.xlabel('Language',fontsize=10)
plt.ylabel('Popularity',fontsize=15)
c = ["r",'y','b','g']
plt.bar(x,y,color="y",width=0.4,label="popularity1")
plt.bar(x,z,color="b",width=0.4,label="popularity2")
plt.legend()
plt.show()
```

```python
import numpy as np

x = ['Python','C','Java','C++']
y = [85,70,80,60]
z = [20,40,50,55]

width = 0.2
p = np.arange(len(x))
p1 = [j+ width for j in p]

plt.title('Matplotlib')
plt.xlabel('Language',fontsize=10)
plt.ylabel('Popularity',fontsize=15)

plt.bar(p,y,color="y",width=0.2,label="popularity1")
plt.bar(p1,z,color="b",width=0.2,label="popularity2")

plt.xticks(p+width/2, x)
plt.legend()
plt.show()
```

```python
import numpy as np

x = ['Python','C','Java','C++']
y = [85,70,80,60]
z = [20,40,50,55]

width = 0.2
p = np.arange(len(x))
p1 = [j+ width for j in p]

plt.title('Matplotlib')
plt.xlabel('Language',fontsize=10)
plt.ylabel('Popularity',fontsize=15)

plt.bar(p,y,color="y",width=0.2,label="popularity1")
plt.bar(p1,z,color="b",width=0.2,label="popularity2")

plt.xticks(p+width/2, x, rotation=10)
plt.legend()
plt.show()
```
